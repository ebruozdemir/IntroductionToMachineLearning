```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.arange(0,20, 0.5)
y = np.sin(x)
#Creating graph
plt.plot(x, y)
#Graph features 
plt.title("Sin(x) Graph")      #name of graph
plt.xlabel("x axis")           #name of x-axis
plt.ylabel("y axis")           #name of y-axis
#Drawing Graph
plt.show()                    
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["Java", "Python", "C", "C#", "HTML","JavaScipt"]
y = np.random.randint(200,500,6)
#Creating column chart
plt.bar(x, y)
#Column Chart's Features
plt.title("Programming Language")
plt.xlabel("Languages")
plt.ylabel("Person number")
# Drawing column chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(80)
y = np.random.rand(80)
color = np.random.rand(80)
size = np.random.randint(80,100,80)
#Creating Chart
plt.scatter(x, y, c=color, s=size)
#Chart Features
plt.title("Covid-19 Chart")
plt.xlabel("-")
plt.ylabel("-")
# Drawing Chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(200)
plt.hist(data, bins=50)
plt.title("Histogram")
plt.xlabel("Time")
plt.ylabel("Data Rate")
plt.show()
```

```python
import matplotlib.pyplot as plt
size = [10,50,20,15,5]
plt.pie(size)
plt.title("Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
x = np.arange(0, 10, 0.50)
y = np.arange(0, 10, 0.50)
x, y = np.meshgrid(x, y)
r = np.sqrt(x**3 + y**3)
z = np.sin(r)
#Creating 3D graph
fig = plt.figure()
ax = fig.add_subplot(212, projection='3d')
ax.plot_surface(x, y, z)
ax.set_title("3D Graphics")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.show()
```

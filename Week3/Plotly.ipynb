# Bar Plot

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.figure_factory as ff
```

```python
timesData= pd.read_csv("timesData.csv")
```

```python
timesData.head()
```

```python
#The first five members of 2015 were selected.
year2015 = timesData[timesData.year == 2015].iloc[:5,:] 
year2015
```

```python
#Creating trace
trace = go.Bar(
                x = year2015.university_name,  #name of x-axis
                y = year2015.teaching,           #name of y-axis
                name = "Teaching",               #name of trace
                marker = dict(color = 'rgba(500, 170, 300, 0.3)',   #color,width process
                             line=dict(color='rgb(0,0,0)',width=1.0)),
                text = year2015.country)      #Place to be shown when the mouse is hovered.

data=[trace]
fig = go.Figure(data = data)
iplot(fig)
```

```python
trace2 = go.Bar(
                x = year2015.university_name,  
                y = year2015.research,   
                name = "Research",    
                marker = dict(color = 'rgba(250, 110, 100, 0.5)',   
                              line=dict(color='rgb(0,0,0)',width=1.0)),
                text = year2015.country)   
data = [trace, trace2]
layout = go.Layout(barmode = "group") 
fig = go.Figure(data = data, layout = layout) 
iplot(fig) 
```

```python
#You can also create trace.
trace_x = { 
  'x': year2015.university_name,
  'y': year2015.teaching, 
  'name': 'Teaching', 
  'type': 'bar'  
};
trace_y = {
  'x': year2015.university_name,
  'y': year2015.research, 
  'name': 'Research',   
  'type': 'bar'  
};
data = [trace_x, trace_y]; 
layout = {
  'xaxis': {'title': 'Top 5 universities'},
  'barmode': 'relative',  
  'title': 'Teaching and Research of top 5 universities in 2015'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Scatter Plot

```python
#First 25 elements were selected for 2012, 2013 and 2014.
year2012 = timesData[timesData.year == 2012].iloc[:25,:]
year2013 = timesData[timesData.year == 2013].iloc[:25,:] 
year2014 = timesData[timesData.year == 2014].iloc[:25,:] 
```

```python
#We created a trace for each year.
trace2012 =go.Scatter( 
                    x = year2012.world_rank, 
                    y = year2012.teaching, 
                    mode = "markers",      #Marker mode appears as dots.
                    name = "2012",       
                    marker = dict(color = 'rgba(300, 118, 225, 0.9)'),
                    text= year2012.university_name) 

trace2013 =go.Scatter(
                    x = year2013.world_rank,
                    y = year2013.teaching,  
                    mode = "markers",      
                    name = "2013",         
                    marker = dict(color = 'rgba(200, 338, 123, 1.6)'),  
                    text= year2013.university_name)  

trace2014 =go.Scatter( 
                    x = year2014.world_rank, 
                    y = year2014.teaching, 
                    mode = "markers",     
                    name = "2014",       
                    marker = dict(color = 'rgba(190, 34, 15, 0.3)'), 
                    text= year2014.university_name) 

data=[trace2012,trace2013,trace2014]
layout = dict(title = 'Teaching and world rank of top 25 universities with 2012, 2013 and 2014 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Teaching',ticklen= 5,zeroline= False)  
             )
fig = dict(data = data, layout = layout)  
iplot(fig) 
```

```python
trace_2012 = { 
  'x': year2012.world_rank,
  'y': year2012.teaching, 
  'name': 'Teaching', 
  'type': 'scatter'  
};
trace_2013 = {
  'x': year2013.world_rank,
  'y': year2013.teaching, 
  'name': 'Teaching',   
  'type': 'scatter'  
};
trace_2014 = {
  'x': year2014.world_rank,
  'y': year2014.teaching, 
  'name': 'Teaching',   
  'type': 'scatter'  
};
data = [trace_2012, trace_2013,trace_2014]; 
layout = {
  'xaxis': {'title': 'World Rank'},
  'yaxis': {'title':'Teaching'}
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Scatter Plot Matrix

```python
dataframe = timesData[timesData.year == 2016]
```

```python
data2016 = dataframe.loc[:,["teaching","research", "income"]] 
```

```python
data2016["index"] = np.arange(1,len(data2016)+1) 
```

```python
data2016.head()
```

```python
fig = ff.create_scatterplotmatrix(
                                  data2016, 
                                  diag='box',  
                                  index='index', 
                                  colormap='Viridis', 
                                  colormap_type='cat',  #cat is categorical
                                  height=500, width=500) 
iplot(fig) 
```

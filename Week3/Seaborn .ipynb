```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#Shows the name of the datasets to be used with Seaborn.
sns.get_dataset_names()
```

```python
#We can load datasets
tips=sns.load_dataset("tips")
iris=sns.load_dataset("iris")
flights=sns.load_dataset("flights")
titanic=sns.load_dataset("titanic")
penguins=sns.load_dataset("penguins")
planets=sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips) #x-axis name, y-axis name, dataset name
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="time") #according to time
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="time", size="size",palette="Pastel1_r")
```

```python
sns.histplot(tips['tip'],bins=10)  #bins shows how many column
```

```python
sns.histplot(tips['tip'],kde=True,bins=20)  #kde shows curve.
```

```python
sns.displot(tips['tip'],kde=True,bins=20)
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette="PuOr")
```

```python
sns.boxplot(x="day",y="tip",data=tips,hue="sex",palette="PuOr")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu", dodge=True) #Dodge separates them from each other.
```

```python
#Facetgrid
g=sns.FacetGrid(tips,col = "smoker")
g=g.map(sns.histplot,"time",color="m")
```

```python
g=sns.FacetGrid(tips,col = "time",row="smoker")
g=g.map(sns.histplot,"total_bill",color="r")
```

```python
sns.set_theme(style="white")
g = sns.JointGrid(data=penguins, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot,
             fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=100, cmap="rocket")
g.plot_marginals(sns.histplot, color="#03051A", alpha=1, bins=20)
```

```python
#We can also create your own dataframe
data = pd.DataFrame({
    'time_zone': np.repeat(['8-10', '11-13', '14-16', '17-19'], 50),
    'rating': np.random.randint(2, 8, 200)
})
g = sns.FacetGrid(data, col="time_zone", margin_titles=True)
g.map(sns.histplot, "rating",color="c")
```
